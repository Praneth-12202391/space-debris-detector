```python
from ultralytics import YOLO

# Load the pretrained model
model = YOLO("yolov8n.pt")

# Train the model
model.train(
    data="/content/data.yaml",
    epochs=50,
    imgsz=640,
    name="space-debris"
)
```

```python
# Evaluate performance
metrics = model.val()
print(metrics)

# Extract metrics properly (no parentheses!)
map_50 = metrics.box.map50          # mAP at IoU=0.5
precision = metrics.box.mp          # mean precision
recall = metrics.box.mr             # mean recall

# Print all metrics
print(f" mAP@0.5: {map_50:.2%}")
print(f" Precision: {precision:.2%}")
print(f" Recall: {recall:.2%}")
```

```python
import cv2
import numpy as np
import onnxruntime
import matplotlib.pyplot as plt
from ultralytics import YOLO
from google.colab import files
from PIL import Image
import io

# -------- Upload Image -------- #
uploaded = files.upload()
img_path = list(uploaded.keys())[0]

# -------- Load Image -------- #
image = Image.open(io.BytesIO(uploaded[img_path]))
original_img = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2BGR)
img_onnx = original_img.copy()
img_combined = original_img.copy()

# -------- Class Names -------- #
class_names = ['large_debris', 'medium_debris', 'rocket', 'satellite', 'small_debris',
               'cheops', 'lisa_pathfinder', 'proba_2', 'smart_1', 'soho', 'xmm_newton']

# -------- ONNX Inference (White Boxes) -------- #
session = onnxruntime.InferenceSession("runs/detect/space-debris/weights/best.onnx")
input_name = session.get_inputs()[0].name

img_resized = cv2.resize(img_onnx, (640, 640))
img_input = img_resized.transpose(2, 0, 1)[None].astype(np.float32) / 255.0
outputs = session.run(None, {input_name: img_input})[0][0]

# Scale factor (in case original image isn't 640x640)
h_scale = original_img.shape[0] / 640
w_scale = original_img.shape[1] / 640

for det in outputs:
    conf = float(det[4])
    if conf < 0.25:
        continue
    x1, y1, x2, y2 = map(int, det[:4])
    x1 = int(x1 * w_scale)
    x2 = int(x2 * w_scale)
    y1 = int(y1 * h_scale)
    y2 = int(y2 * h_scale)
    cls_id = int(det[5])
    if cls_id >= len(class_names):
        continue
    label = f"{class_names[cls_id]} {conf:.2f}"
    cv2.rectangle(img_combined, (x1, y1), (x2, y2), (255, 255, 255), 2)
    cv2.putText(img_combined, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX,
                0.5, (255, 255, 255), 1, cv2.LINE_AA)

# -------- YOLOv8 Inference (Pink Boxes) -------- #
model = YOLO("runs/detect/space-debris/weights/best.pt")
results = model.predict(source=original_img, conf=0.25, save=False)

for r in results:
    for box in r.boxes:
        cls_id = int(box.cls[0])
        conf = float(box.conf[0])
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        name = model.names[cls_id] if cls_id in model.names else "unknown"
        label = f"{name} {conf:.2f}"
        cv2.rectangle(img_combined, (x1, y1), (x2, y2), (255, 102, 255), 2)
        cv2.putText(img_combined, label, (x1, y2 + 15), cv2.FONT_HERSHEY_SIMPLEX,
                    0.5, (255, 102, 255), 1, cv2.LINE_AA)

# -------- Resize for display -------- #
display_img = cv2.resize(img_combined, (600, 600))

# -------- Show Result -------- #
plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(display_img, cv2.COLOR_BGR2RGB))
plt.title("Detection Results")
plt.axis("off")
plt.show()
```

```python
from ultralytics import YOLO
model = YOLO("/content/best.pt")  # or "/content/best.pt" if needed
```

```python
import cv2
import numpy as np

def detect(image):
    results = model.predict(image, conf=0.25)
    img = image.copy()

    for r in results:
        for box in r.boxes:
            cls_id = int(box.cls[0])
            conf = float(box.conf[0])
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            label = f"{model.names[cls_id]} {conf:.2f}"

            cv2.rectangle(img, (x1, y1), (x2, y2), (255, 102, 255), 2)
            cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX,
                        0.5, (255, 102, 255), 2)

    return img
```

```python
import gradio as gr

interface = gr.Interface(fn=detect,
                         inputs=gr.Image(type="numpy", label="Upload Image"),
                         outputs=gr.Image(type="numpy", label="Detected Output"),
                         title="Space Debris Detector",
                         description="Upload an image and detect space debris using YOLOv8!")

interface.launch(share=True)  # share=True gives you a public link
```
